## <span style="color: #9C68C8;">Functions</span>

A **function** is a reusable block of code that performs a specific task and can be called whenever needed in a program. Functions allow you to organize code, reduce repetition, and make your code more readable and efficient.

As we are in the Python AI course, it would be interesting to know in how functions come into play in Artificial Intelligence: AI models resemble functions because, like functions, they take input data, process it, and produce an output based on that data. However, unlike regular functions where the logic is explicitly defined, AI models “learn” the best rules or patterns from data to make predictions or decisions on new inputs.

````python
def greet_user(name, hour):
    if hour < 12:
        greeting = f"Good morning, {name}!"
    elif hour < 18:
        greeting = f"Good afternoon, {name}!"
    else:
        greeting = f"Good evening, {name}!"
    return greeting
````

````python
def convert_temperature(value, unit):
    if unit == 'C':
        return (value * 9/5) + 32  # Convert Celsius to Fahrenheit
    elif unit == 'F':
        return (value - 32) * 5/9  # Convert Fahrenheit to Celsius
    else:
        return "Invalid unit. Please use 'C' for Celsius or 'F' for Fahrenheit."
````
