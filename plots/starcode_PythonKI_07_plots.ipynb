```python
import matplotlib.pyplot as plt
import numpy as np
import base64

np.random.seed(0)
```

```python
language = 'german' # german or english
```

```python
colors = ['#EB6ECB', '#9C68C8', '#6162C5']

if language == 'german':
    x_axis_caption = 'Länge Kronblätter (cm)'
    y_axis_caption = 'Breite Kronblätter (cm)'
    cluster_1_caption = 'Cluster 1'
    cluster_2_caption = 'Cluster 2'
    cluster_3_caption = 'Cluster 3'
    title_text = 'Bildung von "Clustern" bei der Klassifikation von Blättern von Schwertlilien'
    filepath = 'PythonKI_07_clusters_german.png'
else:
    x_axis_caption = 'length of petal leaf (cm)'
    y_axis_caption = 'width of petal leaf (cm)'
    cluster_1_caption = 'Cluster 1'
    cluster_2_caption = 'Cluster 2'
    cluster_3_caption = 'Cluster 3'
    title_text = 'Formation of "Clusters" in the Categorization of Iris Flowers'
    filepath = 'PythonKI_07_clusters_english.png'

center_cluster_1 = np.array([[0],[0]])
circle1 = plt.Circle((0,0), 2.05, ls='--', fill=False, color=colors[0])
center_cluster_2 = np.array([[4],[3]])
circle2 = plt.Circle((4,3), 2.05, ls='--', fill=False, color=colors[1])
center_cluster_3 = np.array([[6],[7]])
circle3 = plt.Circle((6,7), 2.05, ls='--', fill=False, color=colors[2])

cluster_1 = np.random.uniform(low=-1, high=1, size=(2,60)) + 0.35*np.random.randn(2, 60)
cluster_1 = np.add(cluster_1, center_cluster_1)
cluster_2 = np.random.uniform(low=-1, high=1, size=(2,60)) + 0.35*np.random.randn(2, 60)
cluster_2 = np.add(cluster_2, center_cluster_2)
cluster_3 = np.random.uniform(low=-1, high=1, size=(2,60)) + 0.35*np.random.randn(2, 60)
cluster_3 = np.add(cluster_3, center_cluster_3)

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(8,4))

ax1.scatter(cluster_1[0,:], cluster_1[1,:], color='k', marker='x')
ax1.scatter(cluster_2[0,:], cluster_2[1,:], color='k', marker='x')
ax1.scatter(cluster_3[0,:], cluster_3[1,:], color='k', marker='x')

ax2.scatter(cluster_1[0,:], cluster_1[1,:], color=colors[0], marker='x')
ax2.scatter(cluster_2[0,:], cluster_2[1,:], color=colors[1], marker='x')
ax2.scatter(cluster_3[0,:], cluster_3[1,:], color=colors[2], marker='x')

ax2.add_patch(circle1)
ax2.add_patch(circle2)
ax2.add_patch(circle3)

ax2.text(x=-0.5, y=2.5, s='Cluster 1', ha='center')
ax2.text(x=4.5, y=0.5, s='Cluster 2', ha='left')
ax2.text(x=3.75, y=7, s='Cluster 3', ha='right')

ax2.axis('equal')
ax2.set_xlabel(x_axis_caption)
ax1.set_xlabel(x_axis_caption)
ax1.set_ylabel(y_axis_caption)

plt.suptitle(title_text)
plt.savefig(filepath)
plt.show()
```

## Encode image to base64 to implement it in the Jupyter notebook

```python
with open(filepath, "rb") as image_file:
    encoded_string = base64.b64encode(image_file.read())

encoded_string = b'data:image/png;base64,' + encoded_string
# Copy the string to clipboard
from tkinter import Tk
r = Tk()
r.withdraw()
r.clipboard_clear()
r.clipboard_append(encoded_string)
r.update() # now it stays on the clipboard after the window is closed
r.destroy()
```
